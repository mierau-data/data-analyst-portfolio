**Описание признаков**

   - **Name** — название игры.

   - **Platform** — платформа, на которой выпущена игра.

   - **Year_of_Release** — год выпуска игры.

   - **Genre** — жанр игры.

   - **NA_sales** — объём продаж в Северной Америке (в миллионах копий).

   - **EU_sales** — объём продаж в Европе (в миллионах копий).

   - **JP_sales** — объём продаж в Японии (в миллионах копий).

   - **Other_sales** — объём продаж в других странах (в миллионах копий).

   - **Critic_Score** — оценка критиков (от 0 до 100).

   - **User_Score** — оценка пользователей (от 0 до 10).

   - **Rating** — возрастной рейтинг, присвоенный организацией **ESRB (Entertainment Software      Rating Board)**, которая классифицирует видеоигры по возрастным категориям.
   

**Категории рейтинга ESRB**

Организация ESRB (Entertainment Software Rating Board) присваивает видеоиграм возрастные категории в зависимости от содержания.

**Основные категории**

- **RP (Rating Pending)** — рейтинг ещё не присвоен. Используется в рекламных материалах до официальной классификации.
Действует с 1994 года.

- **EC (Early Childhood)** — игры для дошкольников.
Не содержат материалов, которые родители могли бы считать неприемлемыми.
Использовался с 1994 по 2018 год.

- **E (Everyone)** — игры для всех возрастов.
Допускается минимальное насилие и лёгкая разговорная лексика.
До 1998 года обозначался как K-A (Kids to Adults).
Действует с 1994 года.

- **E10+ (Everyone 10 and older)** — игры для пользователей от 10 лет.
Возможны более выраженное насилие и умеренно грубая речь.
Введён в 2005 году.

- **T (Teen)** — игры для подростков от 13 лет.
Могут содержать умеренное насилие (включая небольшое количество крови), грубую лексику, юмор и умеренный сексуальный контент.
Действует с 1994 года.

- **M (Mature)** — игры для лиц от 17 лет.
Возможны реалистичное насилие, сцены крови, ужасы, сексуальный контент, частичная нагота и ненормативная лексика.
Действует с 1994 года.

- **AO (Adults Only)** — игры только для взрослых от 18 лет.
Содержат материалы более высокого уровня воздействия, чем категория M.
Действует с 1994 года.

```python
# Год важен для анализа, поэтому удаляем строки с пропущенным годом, чтобы не искажать результат
df = df.dropna(subset=['year_of_release'])
```

```python
# Теперь приводим год к целочисленному типу
df['year_of_release'] = df['year_of_release'].astype('Int64')
```

```python
# Пропуски в eu_sales, jp_sales, заменяем на средние значения по платформе и году выпуска:  
sales_cols = ['eu_sales', 'jp_sales']
for col in sales_cols:
    df[col] = df[col].fillna(
        df.groupby(['platform', 'year_of_release'])[col].transform('mean'))
```

```python
# Заменяем пропуски в rating на индикатор "no_rating"
df['rating'] = df['rating'].fillna('no_rating')
```

```python
df.isna().sum()
```

Оценки оставлены без заполнения для сохранения корректности анализа.

## Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
df_actual['user_score_category'] = pd.cut(
    df_actual['user_score'],
    bins=[-1, 0, 3, 8, 10],
    labels=['без рейтинга', 'низкая', 'средняя', 'высокая'],
    right=False,
    include_lowest=True
)
```

```python
df_actual['user_score_category'].value_counts()
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
df_actual['critic_score_category'] = pd.cut(
    df_actual['critic_score'],
    bins=[-1, 0, 30, 80, 100], 
    labels=['без рейтинга', 'низкая', 'средняя', 'высокая'],
    right=False,
    include_lowest=True)
```

```python
df_actual['critic_score_category'].value_counts()
```

- После категоризации данных проверьте результат: сгруппируйте данные по выделенным категориям и посчитайте количество игр в каждой категории.

```python
# Группировка по пользовательским категориям
df_actual.groupby('user_score_category')['name'].count()
```

```python
# Группировка по категориям критиков
df_actual.groupby('critic_score_category')['name'].count()
```
